```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot
from sklearn.linear_model import LinearRegression
```

```python
old_data = pandas.read_csv("top-500-movies.csv")
```

```python
old_data.head()
```

```python
data = old_data[['production_cost','worldwide_gross']]
```

```python
data.head()
```

```python
data.describe()
```

```python
x = DataFrame(data,columns=['production_cost'])
y = DataFrame(data,columns=['worldwide_gross'])
```

```python
matplotlib.pyplot.scatter(x,y)
matplotlib.pyplot.show()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x,y)
```

```python
regression.coef_ #theta_1
```

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,alpha=0.3)
plt.plot(x,regression.predict(x),color='black',linewidth=4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression.score(x,y)
```
